```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')

df.head()
```

```python
palette_color = sns.color_palette('Paired')

plt.pie(x=df.groupby('workout_type').workout_type.agg('count'),
        labels = df.groupby('workout_type').workout_type.agg('count').index, colors=palette_color, autopct='%.0f' )
plt.show()
```

Mosaic plots can show you individual days for each group where pie and bar charts will show a total group
