A. ) Store the data as a pandas dataframe. Examine each datatype and comment on the appropriateness of
each.

```python
import pandas as pd
import re
import numpy as np

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')

df.head(20)
```

B.) Remove the column that repeats the indexes and is ‘unnamed’ as a column

```python
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

C.) Clean the column names to remove the unit declaration, (%), using regular expressions. The column
name should not have any trailing spaces after cleaning. You will only receive partial credit for simply renaming
columns without using regular expressions.

```python
df = df.rename(columns=lambda x: re.sub(r'[(.%)]','',x))
```

```python
df.columns
```

D.) Convert ‘activity day’ column into a datetime format.


```python
df['activity_day'].dtype
```

```python
df['activity_day'] = pd.to_datetime(df['activity_day'])

df['activity_day'].dtype
```

E.) Impute ‘total steps’ by replacing the ‘NaN’s with an appropriate number of steps. Convert to appropriate datatype.

```python
df['total_steps'].dtype
```

```python
df['total_steps'].fillna(value=df['total_steps'].min(0), inplace=True)
df.head()
```

F.) Convert non-empty ‘avg pace’ into a float that is still representative of the information contained in the column. Recall that there are 60 seconds in one minute so 3:30 is equivalent to 3.5 minutes.

```python
df['avg_pace'].dtype
```

```python
df['avg_pace'] = pd.DatetimeIndex(df['avg_pace'])
df['avg_pace'] = df['avg_pace'].apply(lambda avg_pace: avg_pace.hour + (avg_pace.minute / 60))
df['avg_pace'].head(10)
```

G.) Group data by ‘workout type’ and find the mean, median, count and standard deviation of ‘calories’.



```python
df.groupby('workout_type').agg(['mean', 'median', 'count', 'std'])['calories']
```

H.)  Create an indicator column that identifies if the activity achieved 30% or more ‘aerobic’ activity

```python
df['aerobic_indicator'] = np.where(df['aerobic'] >= 30, 'True', 'False')
df.head()
```

I.) Which day of the week (Monday, Tuesday, etc.) and ‘workout type‘ has the maximum of the ‘max cadence’.

```python
df1 = df.loc[df['max_cadence'] == df['max_cadence'].max()]
print('maximum_cadence = ' + str(df['max_cadence'].max()))
print('workout_type = ' + df1.workout_type.iloc[0])
print('day_of_week = ' + df1.activity_day.dt.day_name().iloc[0])
```
